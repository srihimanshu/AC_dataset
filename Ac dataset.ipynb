```python
import pandas as pd
import seaborn as sns
```

```python
# uploading dataset
ac_dataset=pd.read_csv("AC_Data.csv")
```

```python
ac_dataset.head()
```

```python
#ac_dataset = df.rename({'AC 1': 'A', 'AC 2': 'B'}, axis=0) for changing the column name 
```

```python
ac_dataset.isnull().sum()
```

```python
ac_dataset["0"]=pd.to_datetime(ac_dataset["0"])
```

```python
#separating date time
ac_dataset["0"]
```

```python
ac_dataset["year"]=ac_dataset["0"].dt.year
ac_dataset["month"]=ac_dataset["0"].dt.month
ac_dataset["day"]=ac_dataset["0"].dt.day
```

```python
ac_dataset["time"]=ac_dataset["0"].dt.time
ac_dataset["hour"]=ac_dataset["0"].dt.hour
```

```python
ac_dataset
```

```python
#df['hour'] = df.index
```

```python
#df["AC 1"].mean()
```

```python
# taking mean hourly
ac_dataset_hour = ac_dataset.groupby(ac_dataset['hour']).mean()
```

```python
ac_dataset_hour
```

```python
ac_dataset_hour=ac_dataset_hour.reset_index()
```

```python
ac_dataset_hour
```

```python
ac_dataset_hour.plot(x='hour', y=[ 'AC 1', 'AC 2', 'AC 3', 'AC 4', 'AC 5', 'AC 6', 'AC 7', 'AC 8',
       'AC 9', 'AC 10', 'AC 11', 'AC 12', 'AC 13', 'AC 14', 'AC 15', 'AC 16',
       'AC 17', 'AC 18'],figsize=(20,10),fontsize=15)
```

## ac power trend shows that it is consuming less power during day time but power consumption increases after 8pm

```python
ac_dataset_hour.plot(x="hour", y='AC 1')
```
